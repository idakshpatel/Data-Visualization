```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and 
#y-values as the sine of x




from bokeh.plotting import figure, output_file, show
import numpy as np

# Create the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Output file
output_file("sine_wave.html")

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the results
show(p)
```

```python
'''2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns'''




from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
import numpy as np
from bokeh.transform import factor_cmap

# Generate data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.randint(10, 50, 50)
colors = np.random.choice(['red', 'blue', 'green', 'purple'], 50)

# Prepare the data for Bokeh
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Output file
output_file("scatter_plot.html")

# Create the plot
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add circle glyphs
p.circle('x', 'y', size='sizes', color='colors', source=source, legend_field="colors", fill_alpha=0.6)

# Show the plot
show(p)
```

```python
'''3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits - ['Apples', 'Oranges',
'Bananas', 'Pears']
counts =[20, 25, 30, 35]

'''
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap

# Create the data
fruits = ['Apples', 'Pears', 'Oranges', 'Bananas']
counts = [20, 25, 30, 35]

# Prepare the data for Bokeh
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Output file
output_file("bar_chart.html")

# Create the plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', plot_height=400, plot_width=600)

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_field="fruits", 
       line_color='white', fill_color=factor_cmap('fruits', palette=['red', 'green', 'orange', 'yellow'], factors=fruits))

# Show the plot
show(p)
```

```python
'''4. Create a Bokeh histogram to visualize the distribution Of the given data.

data_hist =np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

'''


from bokeh.plotting import figure, output_file, show
import numpy as np

# Generate data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Output file
output_file("histogram.html")

# Create a new plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph to represent the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```

```python
'''
5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = IO)
x = np.1inspace(O, 1, 10)
y :: np.Iinspace(O, I, IO)
xx, yy = np.meshgrid(x, y)

'''


from bokeh.plotting import figure, output_file, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
import numpy as np
from bokeh.palettes import Viridis256

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Output file
output_file("heatmap.html")

# Convert the data into a flat format suitable for Bokeh
heatmap_data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten(),
}

# Set up the plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', plot_width=600, plot_height=600, toolbar_location=None, tools="")

# Define a color mapper for the heatmap
mapper = LinearColorMapper(palette=Viridis256, low=min(heatmap_data['value']), high=max(heatmap_data['value']))

# Add rect glyphs to create the heatmap
p.rect(x='x', y='y', width=1, height=1, source=heatmap_data, fill_color=transform('value', mapper), line_color=None)

# Add a color bar to the right of the plot
color_bar = ColorBar(color_mapper=mapper, location=(0, 0), ticker=None)
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
