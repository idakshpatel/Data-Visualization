```python
'''1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic 
dataset.'''

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 1. Scatter Plot with a Synthetic Dataset
np.random.seed(42)
x = np.random.randn(100)  # 100 random points for variable X
y = 2.5 * x + np.random.randn(100)  # Y is correlated with X with added noise

plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', alpha=0.7)
plt.title('Scatter Plot of Synthetic Dataset')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.grid(True)
plt.show()

```

```python
#2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

# Distribution of a Numerical Variable
data = np.random.normal(50, 15, 1000)  # Generating 1000 random numbers from a normal distribution

plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, color='green', kde=True)  # KDE adds a smooth curve representing the distribution
plt.title('Distribution of Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
'''3. Create a dataset representing categories and their corresponding values. Compare different categories 
based on numerical values.'''

#Categories and Corresponding Values Comparison (Bar Plot)
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [40, 70, 30, 85, 55]

plt.figure(figsize=(10, 5))
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
'''4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
variable across different categories.'''



#Distribution of Numerical Variable Across Categories (Box Plot)
np.random.seed(42)
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, 200),
    'Value': np.random.normal(50, 10, 200)
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 5))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
'''5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
heatmap'''


#Correlation Matrix Visualization (Heatmap)
np.random.seed(42)
data = np.random.randn(100, 5)  # Generating a dataset with 5 correlated features
df_corr = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4', 'Feature 5'])
corr_matrix = df_corr.corr()  # Calculating the correlation matrix

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
